# Cardio Good Fitness Case Study - Descriptive Statistics
##### The market research team at AdRight is assigned the task to identify the profile of the typical customer for each treadmill product offered by CardioGood Fitness. The market research team decides to investigate whether there are differences across the product lines with respect to customer characteristics. The team decides to collect data on individuals who purchased a treadmill at a CardioGoodFitness retail store during the prior three months. The data are stored in the CardioGoodFitness.csv file.
### The team identifies the following customer variables to study:
##### <li>product purchased, TM195, TM498, or TM798 <li> gender <li> age, in years <li> education, in years <li> relationship status, single or partnered <li> annual household income ($) <li> average number of times the customer plans to use the treadmill each week <li> average number of miles the customer expects to walk/run each week <li> and self-rated fitness on an 1-to-5 scale, where 1 is poor shape and 5 is excellent shape

# Import Necessary Packages

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

# Loading Data

```python
data = pd.read_csv("CardioGoodFitness.csv")
data.head()
```

# Analysis

```python
data.describe(include="all")
```

## Checking for Null Values

```python
data.info()
```

```python
data.isnull().sum()
```

## Count Plots

```python
fig, axes = plt.subplots(figsize=(30, 15), nrows=2, ncols=3)

sns.countplot(x=data.Product, ax=axes[0, 0])
axes[0, 0].set_title('Product Sales')
axes[0, 0].set_xlabel('Product')
axes[0, 0].set_ylabel('Quantity Sold')

sns.countplot(x=data.Gender, ax=axes[0, 1])
axes[0, 1].set_title('Gender Count')
axes[0, 1].set_xlabel('Gender')
axes[0, 1].set_ylabel('Count')

sns.countplot(x=data.Education, ax=axes[0, 2])
axes[0, 2].set_title('Education Count')
axes[0, 2].set_xlabel('Education')
axes[0, 2].set_ylabel('Count')

sns.countplot(x=data.MaritalStatus, ax=axes[1, 0])
axes[1, 0].set_title('MaritalStatus Count')
axes[1, 0].set_xlabel('MaritalStatus')
axes[1, 0].set_ylabel('Count')

sns.countplot(x=data.Usage, ax=axes[1, 1])
axes[1, 1].set_title('Usage Count')
axes[1, 1].set_xlabel('Usage')
axes[1, 1].set_ylabel('Count')

sns.countplot(x=data.Fitness, ax=axes[1, 2])
axes[1, 2].set_title('Fitness Count')
axes[1, 2].set_xlabel('Fitness')
axes[1, 2].set_ylabel('Count')

# set the spacing between subplots
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.5, 
                    top=0.5, 
                    wspace=0.4, 
                    hspace=0.4)
```

## Distributon/Histogram Plots

```python
fig, axes = plt.subplots(figsize=(20, 10), nrows=2, ncols=2)

sns.histplot(data.Income, ax=axes[0, 0])
axes[0, 0].set_title('Income Distribution')
axes[0, 0].set_xlabel('Income')
axes[0, 0].set_ylabel('Density')

sns.histplot(data.Miles, ax=axes[0, 1])
axes[0, 1].set_title('Miles Walked Distribution')
axes[0, 1].set_xlabel('Miles Walked')
axes[0, 1].set_ylabel('Density')

sns.kdeplot(data.Income, ax=axes[1, 0])
axes[1, 0].set_title('Income Distribution')
axes[1, 0].set_xlabel('Income')
axes[1, 0].set_ylabel('Density')

sns.kdeplot(data.Miles, ax=axes[1, 1])
axes[1, 1].set_title('Miles Walked Distribution')
axes[1, 1].set_xlabel('Miles Walked')
axes[1, 1].set_ylabel('Density')
```

## Scatter Plots

```python
plt.rcParams["figure.figsize"] = (10,8)
sns.scatterplot(x="Age", y="Income", hue="Gender", data=data)
plt.title("Age vs Income")
plt.show()
```

```python
sns.scatterplot(x="Age", y="Miles", hue="Gender", data=data)
plt.title("Age vs Miles Walked")
plt.ylabel("Miles Walked")
plt.xlabel("Age (in yrs)")
plt.show()
```

## Box Plots

```python
fig, axes = plt.subplots(figsize=(20, 8), nrows=1, ncols=3)

sns.boxenplot(x='Gender',y='Age',data=data, ax=axes[0])
axes[0].set_title("Gender vs Age")

sns.boxenplot(x='Gender',y='Income',data=data, ax=axes[1])
axes[1].set_title("Gender vs Income")

sns.boxenplot(x='Gender',y='Miles',data=data, ax=axes[2])
axes[2].set_title("Gender vs Miles")

plt.show()
```

```python
fig, axes = plt.subplots(figsize=(20, 8), nrows=1, ncols=3)

sns.boxenplot(x='Product',y='Age',data=data, ax=axes[0])
axes[0].set_title("Product vs Age")

sns.boxenplot(x='Product',y='Income',data=data, ax=axes[1])
axes[1].set_title("Product vs Income")

sns.boxenplot(x='Product',y='Miles',data=data, ax=axes[2])
axes[2].set_title("Product vs Miles")

plt.show()
```

## Corelation Heat Map

```python
sns.heatmap(data.corr(), annot=True)
plt.show()
```

## Pair Plot

```python
sns.pairplot(data, hue="Gender")
plt.title("Pair Plot")
plt.show()
```
